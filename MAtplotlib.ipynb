```python
Children=[10, 25, 15, 30, 20] # Added sample data
color=['red','blue','green','yellow','pink']
plt.bar(color,Children)
plt.xlabel('Color') # Added x-axis label
plt.ylabel('Number of Children') # Added y-axis label
plt.title('Number of Children by Color') # Added title
plt.show()
```

#**Multiple Bar chart**

```python
df=pd.read_csv('batsman_season_record.csv')
print(df)
plt.bar(df['batsman'],df['2015'])
plt.show()
```

```python
import numpy as np

plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

## **Advance Matplotlib**

```python
iris=pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.show()
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

# **changing graph size**

```python
plt.figure(figsize=(15,10)) #this is for size of figure

#lower code is decribe the figure
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
```
