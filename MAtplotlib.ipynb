# **Annotation**

```python
batters=pd.read_csv('batter.csv')
```

```python
batters.shape
```

```python
sample_df=batters.head(100).sample(25,random_state=5)
```

```python
print(sample_df)
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate']) # this is only sccaters plot of batsman
```

```python
plt.figure(figsize=(11,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate']) # this is only sccaters plot of batsman
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) # this is only sccaters plot of batsman
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) # this is only sccaters plot of batsman
plt.axhline(130,color='indigo')
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) # this is only sccaters plot of batsman
plt.axhline(130,color='darkblue')
plt.axvline(130,color='darkblue')
for i in range (sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## **subplots**

**A difference way to plot graph**

```python
plt.scatter(batter['avg'],batter['strike_rate'])
plt.title("something")
```

```python
# changing size of graph
plt.figure(figsize=(10,5))
# batter
fig,ax=plt.subplots()
ax.scatter(batter['avg'],batter['strike_rate'])
# adding labels
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show
```

```python

# changing size of graph
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(batter['avg'],batter['strike_rate'])
# adding tittle in it
ax.set_title("something")
# adding labels
# adding labels
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
# ploting two plots (subplots)
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(30,20),sharex=True)
plt.title("difference bitween avg and strike_rate and avg and runs")
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0].set_title("difference bitween avg and strike_rate ")
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')
ax[1].scatter(batter['avg'],batter['runs'])
ax[1].set_title("difference bitween avg and runs ")
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
fig.show()
```

```python
# ploting 1 plots (4 subplots)
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(30,20))
plt.title("difference bitween avg and strike_rate and avg and runs")
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0,0].set_title("difference bitween avg and strike_rate ")
ax[0,0].set_xlabel('Average')
ax[0,0].set_ylabel('Strike Rate')
ax[1,0].scatter(batter['avg'],batter['runs'])
ax[1,0].set_title("difference bitween avg and runs ")
ax[1,0].set_xlabel('Average')
ax[1,0].set_ylabel('Runs')
ax[0,1].hist(batter['avg'])
ax[0,1].set_title("Histogram of Average")
ax[0,1].set_xlabel('Average')
ax[1,1].hist(batter['strike_rate'])
ax[1,1].set_title("Histogram of Strike Rate")
ax[1,1].set_xlabel('Strike Rate')
fig.show()
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'])

ax2=fig.add_subplot(2,2,2)
ax2.scatter(batter['runs'],batter['strike_rate'])

ax3=fig.add_subplot(2,2,3)
ax3.scatter(batter['avg'],batter['runs'])

ax4=fig.add_subplot(2,2,4)
ax4.scatter(batter['runs'],batter['avg'])
```

# **3d Scatter plot**
